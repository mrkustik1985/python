Реализуйте функцию, которая для каждого объекта тестового набора данных найдет $k$ ближайших соседей из обучающего набора данных, и в качестве ответа возьмет наиболее часто встречающееся значение целевого признака среди найденных соседей. Обратите внимание, что соседей необходимо искать только среди объектов обучающего набора данных.

```python
# при необходимости обучения на большем числе стобцов
def replace_gender_value(ds):
    copy = ds.copy()
    copy.loc[copy["Sex"] == "male", "Sex"] = 1
    copy.loc[copy["Sex"] == "female", "Sex"] = 0
    return copy


train_ft = replace_gender_value(train_ft)
test_ft = replace_gender_value(test_ft)
test_ft
```

```python
# тк от возраста нет зависимости умерших/выживших => все значения NaN положим на эти места среднюю
avrg_train = np.mean(train_ft["Age"])
avrg_test = np.mean(test_ft["Age"])
train_ft_copy = train_ft.copy()
train_ft_copy["Age"] = train_ft_copy["Age"].fillna(avrg_train)
test_ft_copy = test_ft.copy()
test_ft_copy["Age"] = test_ft_copy["Age"].fillna(avrg_test)
train_ft_copy
```

Теперь давайте нормализуем критерии и посмотрим что стало с таблицей(ниже knn написано)

```python
def replace_ds(ds):
    ds = ds[["Age", "Fare"]]
    mean = ds.mean()
    standard = ds.std()
    answer_ds = (ds - mean) / standard
    return answer_ds
```

```python
def knn(features, train_features, train_target, k=16):
    features = np.array(features)
    train_features = np.array(train_features)
    dst = np.sqrt(np.sum((features[:, np.newaxis, :] - train_features) ** 2, axis=2))
    dst = np.argsort(dst, axis=1)[:, :k]
    delta = np.vectorize(lambda i: train_target[i])(dst)
    return pd.Series(
        np.apply_along_axis(lambda x: np.bincount(x).argmax(), axis=1, arr=delta)
    )
```

```python
# функция оценки точности
def definatly(array1, array2):
    array1 = np.array(array1)
    array2 = np.array(array2)
    array3 = array1 == array2
    sm = np.sum(array3)
    return sm / len(array1)
```

В данной задаче достаточно, если первое дерево будет иметь максимальную глубину 2, а второе &mdash; глубину 1, и при этом не является поддеревом первого. *Примечание*: дерево из одного листа имеет глубину 0.

Реализуйте полученные деревья по шаблону

```python
def tree1(train, k1=6, k2=2):
    # k1 - fare, k2 - Pclass
    condition1 = train["Fare"] < k1
    condition2 = train["Pclass"] < k2
    predictions = pd.Series(0, index=train.index)
    predictions[condition1 | condition2] = 1
    return pd.Series(predictions)


def tree2(train, k1=30, k2=0.5):
    # k1 - age, k2 - SibSp
    condition1 = train["Age"] < k1
    condition2 = train["SibSp"] >= k2
    condition3 = train["Parch"] >= 1
    predictions = pd.Series(0, index=train.index)
    predictions[condition1 & condition2 & condition3] = 1
    return pd.Series(predictions)
```

```python
fare = np.linspace(0, 30, 30)
pclass = [0, 1, 2]
accuracies = []
fare_tst = []
pclass_tst = []
for fr in fare:
    for x in pclass:
        predictions_param2_1 = tree1(train_ft, fr, x)
        accur = definatly(train_trgt, predictions_param2_1)
        accuracies.append(accur)
        fare_tst.append(fr)
        pclass_tst.append(x)

plt.scatter(fare_tst, accuracies, c=pclass_tst, cmap="viridis", alpha=0.5)
plt.colorbar(label="pclass")
plt.xlabel("Fare")
plt.ylabel("Accuracy")
plt.title("Fare vs Accuracy with pclass color representation")
plt.show()
```

Реализуйте данный критерий по шаблону

```python
def criteria(target, predicted):
    # аналогичен defenitly - функции точности с округлением до 10**-3
    return (round(definatly(target, predicted) * 10**3)) / 10**3
```
